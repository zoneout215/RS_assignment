```python
from datetime import datetime
from sklearn.metrics.pairwise import cosine_similarity
import json
import pandas as pd
import numpy as np
import os 
```

```python
current_dir = os.getcwd()  # gets current working directory
project_dir = os.path.dirname(current_dir)  
directory_path = os.path.dirname(current_dir) + os.sep + 'data' + os.sep
```

```python
movies = pd.read_json(directory_path + 'movie_activities.json')
shows =  pd.read_json(directory_path + 'show_activities.json')

movies
```

```python
movies.columns
```

```python
watched_movies = movies[movies['activity'] == 'Play']

most_viewed = watched_movies.groupby('content_id', as_index = False).count().sort_values('activity', ascending = False)[:10]

most_viewed['content_id'].reset_index(drop = True).to_csv(directory_path + 'most_viewed_movies.csv', index = False)

pd.read_csv(directory_path + 'most_viewed_movies.csv')
```

```python
watched_shows = shows[shows['activity'] == 'Play']

most_viewed_shows = watched_shows.groupby('content_id', as_index = False).count().sort_values('activity', ascending = False)[:10]

most_viewed_shows['content_id'].reset_index(drop = True).to_csv(directory_path + 'most_viewed_shows.csv', index = False)

pd.read_csv(directory_path + 'most_viewed_shows.csv')
```
