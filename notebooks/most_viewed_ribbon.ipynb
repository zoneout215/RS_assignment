```python
from datetime import datetime
from sklearn.metrics.pairwise import cosine_similarity
import json
import pandas as pd
import numpy as np
import os
```

```python
path = os.getcwd() + os.sep + 'data' + os.sep

activities = pd.read_json(path + 'activities.json')
```

```python
movies = activities[activities['content_id'] < 1000]

shows =  activities[activities['content_id'] > 1000]

movies
```

```python
shows
```

```python
watched_movies = movies[movies['activity'] == 'Play']

most_viewed = watched_movies.groupby('content_id', as_index = False).count().sort_values('activity', ascending = False)[:10]

most_viewed['content_id'].reset_index(drop = True).to_csv(path + 'most_viewed_movies.csv', index = False)

pd.read_csv(path + 'most_viewed_movies.csv')
```

```python
watched_shows = shows[shows['activity'] == 'Play']

most_viewed_shows = watched_shows.groupby('content_id', as_index = False).count().sort_values('activity', ascending = False)[:10]

most_viewed_shows['content_id'].reset_index(drop = True).to_csv(path + 'most_viewed_shows.csv', index = False)

pd.read_csv(path + 'most_viewed_shows.csv')
```
